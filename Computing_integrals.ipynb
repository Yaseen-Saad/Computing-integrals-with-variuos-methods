# Computing Integrals using Various Methods

## I. Derivation of Rules

### 1. The Trapezoidal Rule

#### 1. Introduction

To approximate the integral of a function $f(x)$ over the interval $[a, b]$, we use the **Trapezoidal Rule**. The goal is to compute an approximation for the following integral:

$$
\int_a^b f(x) \, dx.
$$

Instead of directly calculating the area under the curve, which may be complex, we divide the area under $f(x)$ into several **trapezoids** and calculate the area of each.

#### 2. Dividing the Interval into Subintervals

We start by breaking the interval $[a, b]$ into $n$ **equal subintervals**. This gives us $n$ trapezoids to approximate the area.

**Step 1: Define the Width of Each Subinterval**

The width $h$ of each subinterval can be calculated as:

$$
h = \frac{b - a}{n},
$$

where:

- $b - a$ is the total length of the interval,
- $n$ is the number of trapezoids.

Thus, $h$ represents the horizontal distance between the points at which we will evaluate $f(x)$.

**Step 2: Define the $x$-Coordinates for Each Subinterval**

We define specific points $x_i$ along the interval where we will evaluate $f(x)$. These points are as follows:

- **Left endpoint**: $x_0 = a$
- **Right endpoint**: $x_n = b$
- **Intermediate Points**: For each $x_i$ where $i = 1, 2, \ldots, n-1$, the coordinates are calculated as:

$$
x_i = a + i \cdot h.
$$

In total, we get $n + 1$ points: $x_0, x_1, x_2, \ldots, x_n$.

#### 3. Deriving the Area of Each Trapezoid

The idea behind the Trapezoidal Rule is to approximate each small section under $f(x)$ by a **trapezoid**.

**Step 3: Area of a Trapezoid**

The area of any trapezoid with parallel sides is given by:

$$
A = \frac{1}{2}(B_1 + B_2) H,
$$

where $B_1$ and $B_2$ are the lengths of the parallel sides, and $H$ is the height (or width) between them.

Thus, for a trapezoid with parallel sides of lengths $f(x_{i-1})$ and $f(x_i)$ and a base (or width) $h$, the area $A_i$ of the trapezoid can be computed as:

$$
A_i = \frac{1}{2} \left( f(x_{i-1}) + f(x_i) \right) h.
$$

**Step 4: Summing the Areas of All Trapezoids**

To approximate the total area $A$ under the curve from $a$ to $b$, we add up the areas of all $n$ trapezoids:

$$
A \approx \sum_{i=1}^{n} A_i = \sum_{i=1}^{n} \frac{1}{2} \left( f(x_{i-1}) + f(x_i) \right) h.
$$

This summation represents our approximation of the total area.

#### 4. Simplifying the Summation Expression

We can further simplify the expression by reorganizing terms:

1. **Endpoints**: The values at $x_0 = a$ and $x_n = b$ appear only once in the summation.
2. **Intermediate Points**: For each $x_i$ where $1 \leq i \leq n - 1$, the function value $f(x_i)$ appears twice because it contributes to the area of both neighboring trapezoids.

Thus, we can rewrite the summation as follows:

$$
A \approx \frac{h}{2} \left( f(a) + 2 \sum_{i=1}^{n-1} f\left( a + i h \right) + f(b) \right).
$$

**Final Formula for the Trapezoidal Rule**

The Trapezoidal Rule for approximating the integral $\int_a^b f(x) \, dx$ is:

$$
\int_a^b f(x) \, dx \approx \frac{h}{2} \left( f(a) + 2 \sum_{i=1}^{n-1} f\left( a + i h \right) + f(b) \right).
$$

### 2. Simpson's One-Third Rule

#### 1. Introduction

To approximate the integral of a function $f(x)$ over the interval $[a, b]$, we use **Simpson's One-Third Rule**:

$$
\int_a^b f(x) \, dx.
$$

Simpson's Rule improves upon the Trapezoidal Rule by approximating the integrand $f(x)$ with a **parabola** (quadratic polynomial) on each pair of subintervals instead of straight lines. This leads to a more accurate approximation, especially when $f(x)$ is smooth.

#### 2. Dividing the Interval into Subintervals

Simpson's One-Third Rule divides the interval $[a, b]$ into $n$ **equal subintervals**, where $n$ must be an even integer.

**Step 1: Define the Width of Each Subinterval**

The width $h$ of each subinterval is:

$$
h = \frac{b - a}{n},
$$

where $b - a$ is the total length of the interval.

**Step 2: Define the $x$-Coordinates for Each Subinterval**

We define the points $x_i$ along the interval where we will evaluate $f(x)$:

- **Left endpoint**: $x_0 = a$
- **Right endpoint**: $x_n = b$
- **Intermediate Points**: For each $x_i$ where $i = 1, 2, \ldots, n - 1$:

$$
x_i = a + i \cdot h.
$$

#### 3. Approximating $f(x)$ with a Parabola

For each pair of subintervals (from $x_{i-1}$ to $x_{i+1}$), we approximate $f(x)$ by a quadratic polynomial:

$$
f(x) \approx \alpha x^2 + \beta x + \gamma,
$$

where $\alpha$, $\beta$, and $\gamma$ are constants to be determined.

#### 4. Calculating the Area Under the Parabola

The area under the parabola from $x_{i-1}$ to $x_{i+1}$ is:

$$
\int_{x_{i-1}}^{x_{i+1}} f(x) \, dx \approx \int_{x_{i-1}}^{x_{i+1}} (\alpha x^2 + \beta x + \gamma) \, dx = \left[ \frac{\alpha x^3}{3} + \frac{\beta x^2}{2} + \gamma x \right]_{x_{i-1}}^{x_{i+1}}.
$$

#### 5. Relating the Parameters to Function Values

To determine $\alpha$, $\beta$, and $\gamma$, we consider an interval from $-1$ to $+1$ (a standardized interval), and we map our general interval to this standard interval.

**Evaluating at Three Points**

We evaluate $f(x)$ at three points within the interval:

1. At $x = -1$:

   $$
   f(-1) = \alpha (-1)^2 + \beta (-1) + \gamma = \alpha - \beta + \gamma.
   $$

2. At $x = 0$:

   $$
   f(0) = \gamma.
   $$

3. At $x = 1$:

   $$
   f(1) = \alpha (1)^2 + \beta (1) + \gamma = \alpha + \beta + \gamma.
   $$

**Solving for the Coefficients**

Solving these equations for $\alpha$, $\beta$, and $\gamma$:

$$
\begin{align*}
\alpha &= \frac{f(1) + f(-1)}{2} - f(0), \\
\beta &= \frac{f(1) - f(-1)}{2}, \\
\gamma &= f(0).
\end{align*}
$$

#### 6. Expressing the Integral as a Weighted Sum

We calculate the area under the parabola over the interval $[-1, 1]$:

$$
\int_{-1}^{1} (\alpha x^2 + \beta x + \gamma) \, dx = \left[ \frac{\alpha x^3}{3} + \frac{\beta x^2}{2} + \gamma x \right]_{-1}^{1} = \frac{2\alpha}{3} + 2\gamma.
$$

Substituting the expressions for $\alpha$ and $\gamma$:

$$
\int_{-1}^{1} f(x) \, dx \approx \frac{f(-1)}{3} + \frac{4f(0)}{3} + \frac{f(1)}{3}.
$$

Thus, the area under the parabola can be approximated using the values of $f(x)$ at $x = -1, 0, 1$.

#### 7. Generalizing to Any Interval

To apply this result to any interval $[x_{i - h}, x_{i + h}]$, where $h$ is the width of each subinterval, we map the interval $[-1, 1]$ to $[x_{i - h}, x_{i + h}]$.

The approximation becomes:

$$
\int_{x_{i - h}}^{x_{i + h}} f(x) \, dx \approx \frac{h}{3} f_{i - 1} + \frac{4h}{3} f_i + \frac{h}{3} f_{i + 1},
$$

where:

- $f_{i - 1} = f(x_{i - h})$,
- $f_i = f(x_i)$,
- $f_{i + 1} = f(x_{i + h})$.

#### 8. Applying Simpson's Rule Over the Entire Interval

Simpson's Rule requires that the total number of intervals $n$ is even because it operates over pairs of intervals.

The integral over the entire interval $[a, b]$ can be approximated by summing over each pair of intervals:

$$
\int_a^b f(x) \, dx \approx \sum_{i=1}^{n/2} \left( \frac{h}{3} f_{2i - 2} + \frac{4h}{3} f_{2i - 1} + \frac{h}{3} f_{2i} \right).
$$

Combining the terms, we get:

$$
\int_a^b f(x) \, dx \approx \frac{h}{3} \left[ f(x_0) + 4 \sum_{i=1}^{n/2} f(x_{2i - 1}) + 2 \sum_{i=1}^{n/2 - 1} f(x_{2i}) + f(x_n) \right].
$$

**Final Formula for Simpson's One-Third Rule**

The Simpson's One-Third Rule for approximating the integral $\int_a^b f(x) \, dx$ is:

$$
\int_a^b f(x) \, dx \approx \frac{h}{3} \left[ f(a) + 4 \sum_{i=1,\, \text{odd}}^{n-1} f(x_i) + 2 \sum_{i=2,\, \text{even}}^{n-2} f(x_i) + f(b) \right].
$$

- The first sum runs over all odd indices $i$.
- The second sum runs over all even indices $i$.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
from tqdm import tqdm
from scipy.integrate import quad
```

```python
# Define the  function
def f(x):
    return np.exp(-(x**2))
```

```python
ns = [10, 100, 1000, 10000, 100000, 1000000]
a = 0
b = 1
trials = 10
```

#### Recalling the Formulla of Trapezoidal Integration:
$$
\int_a^b f(x) \, dx \approx \frac{h}{2} \left( f(a) + 2 \sum_{i=1}^{n-1} f\left( a + i h \right) + f(b) \right).
$$

```python
def trapezoidal_integration(f, a, b, n):

    xs = np.linspace(a, b, n + 1)

    h = (b - a) / n

    A = h / 2

    S = f(a) + f(b)

    S += 2 * np.sum(f(xs)[1:-1])

    return A * S
```

#### Recalling the Formulla of Simpson's Integration:

$$
\int_a^b f(x) \, dx \approx \frac{h}{3} \left[ f(a) + 4 \sum_{i=1,\, \text{odd}}^{n-1} f(x_i) + 2 \sum_{i=2,\, \text{even}}^{n-2} f(x_i) + f(b) \right].
$$


```python
def simpsons_integration(f, a, b, n):
    if n % 2:
        return "n must be even"
    xs = np.linspace(a, b, n + 1)
    h = (b - a) / n
    A = h / 3
    S = f(a) + f(b)
    S += 4 * np.sum(f(xs)[1:-1:2]) + 2 * np.sum(f(xs)[2:-1:2])
    return A * S
```

```python
# Graph the function to assign the monte de carlo bounderies
# Generate x values from 0 to 1
x_values = np.linspace(0, 1, ns[-1])
y_values = f(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$e^{-x^2}$', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $e^{-x^2}$ from x = 0 to x = 1')
plt.legend()
plt.grid(True)
plt.show()
```

Thus, the boundries are (0,0.4) (1,1)

```python
def monte_carlo_integration(f, a, b, n):
    hits = 0
    for _ in range(1, n):
        x = random.uniform(a, b)
        y = random.uniform(0, 1)
        if y < f(x):
            hits += 1
    return hits / n * 1 * 1
```

```python
traps,simps,montes = [],[],[]
```

```python
for n in tqdm(ns, desc="Processing different n values"):
    trap, simp, monte = [], [], []

    # Inner loop with an explicit tqdm bar
    for i in tqdm(range(1, trials), desc=f"Trials for n={n}", leave=False, position=1):
        trap.append(trapezoidal_integration(f, a, b, n))
        simp.append(simpsons_integration(f, a, b, n))
        monte.append(monte_carlo_integration(f, a, b, n))

    # Append the averaged results
    traps.append(np.mean(trap))
    simps.append(np.mean(simp))
    montes.append(np.mean(monte))
```

```python
exact_result  = quad(f, a, b)[0]
```

```python

# Calculate absolute errors for each method
traps_error = [abs(trap - exact_result) for trap in traps]
simps_error = [abs(simp - exact_result) for simp in simps]
montes_error = [abs(monte - exact_result) for monte in montes]

# Create a new figure for the error plot
plt.figure(figsize=(10, 6))

# Plot the absolute error for each integration method
plt.plot(ns, traps_error, 'o-', label="Trapezoidal Rule Error", markersize=5)
plt.plot(ns, simps_error, 's-', label="Simpson's Rule Error", markersize=5)
plt.plot(ns, montes_error, 'x-', label="Monte Carlo Method Error", markersize=5)

# Configure plot labels, title, and scale
plt.xlabel("Number of Points (n)")
plt.ylabel("Absolute Error")
plt.title("Absolute Error of Integration Methods for Each n")
plt.xscale('log')
plt.yscale('log')  # Log scale on y-axis for better visualization of error magnitudes
plt.grid(True)

# Add a single legend without duplicates
plt.legend()

# Display the plot
plt.show()
```
